# Storytelling with Data Viz on Exchange Rates

## The Data

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021.

Our focus in the guided part of the project will be on the exchange rate between the euro and the American dollar.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate)

```python
import pandas as pd

# import the data into a dataframe

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

### Data Cleaning

```python
# rename to make typing easier, convert 'Time' object to datetime

exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

We focus on US_dollar

```python
# isolate Time and US_dollar columns

euro_to_dollar = exchange_rates[['Time','US_dollar']]

euro_to_dollar.head()
```

```python
# investigate the distribution of unique values

euro_to_dollar['US_dollar'].value_counts()
```

We want to get rid of the character "-".

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

Let's convert the US_dollar column to float:

```python
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)
```

```python
euro_to_dollar.info()
```

#### Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.show()
```

If we want to only show the long-term trends in the exchange rate, we can use the rolling mean (moving average) to hide the daily variations.

To accomplish this, we can use pandas.Series.rolling().mean() method.

Let's calculate the rolling mean for US_dollar using a rolling window of 30 days and let's add this calculation to a new column called rolling_mean.

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(30).mean()
```

```python
euro_to_dollar.head(40)
```

One interesting idea is to investigate how the euro-dollar rate changed under the three previous US presidents (George W. Bush (January 20, 2001 - January 20, 2009), Barack Obama (January 20, 2009 - January 20, 2017), and Donald Trump (January 20, 2017- January 20, 2021)).

Let's code this up!

### The EUR/USD Exchange Rate under Bush, Obama, and Trump

Let's get the data series for the respective term dates.

```python
all_presidents = euro_to_dollar[euro_to_dollar["Time"].between('2001-01-20', '2021-01-20')]
```

```python
all_presidents.head()
```

```python
all_presidents.tail()
```

```python
all_presidents["rolling_mean"].describe()
```

```python
bush = all_presidents[all_presidents["Time"].between('2001-01-20', '2009-01-20')]
```

```python
obama = all_presidents[all_presidents["Time"].between('2009-01-21', '2017-01-20')]
```

```python
trump = all_presidents[all_presidents["Time"].between('2017-01-21', '2021-01-20')]
```

```python
# use FiveThirtyEight style

import matplotlib.style as style

style.use('fivethirtyeight')

# add subplots

plt.figure(figsize = (12, 6))

ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)

ax4 = plt.subplot(2, 1, 2)

axes = [ax1, ax2, ax3, ax4]

# format subplots using statistics from above

for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha = 0.3)
    ax.grid(alpha = 0.5)
    
    
# Colours for the presidents:

# Bush:   #ADE4DB
# Obama:  #F6BA6F
# Trump:  #6DA9E4

# Background: #FFEBEB
    
    
# plot Bush term

ax1.plot(bush["Time"], bush["rolling_mean"], color = '#ADE4DB')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha = 0.3)
ax1.text(12300, 1.92, 'BUSH', fontsize = 18, weight = "bold", color = '#ADE4DB')
ax1.text(12000, 1.80, '(2001 - 2009)', weight = "bold", alpha = 0.3)

# plot Obama term

ax2.plot(obama["Time"], obama["rolling_mean"], color = '#F6BA6F')
ax2.set_xticklabels(['','2009','', '2011', '', '2013', '', '2015', '', '2017'], alpha = 0.3)
ax1.text(16000, 1.92, 'OBAMA', fontsize = 18, weight = "bold", color = '#F6BA6F')
ax1.text(15800, 1.80, '(2009 - 2017)', weight = "bold", alpha = 0.3)

ax3.plot(trump["Time"], trump["rolling_mean"], color = '#6DA9E4')
ax3.set_xticklabels(['2017','', '2018', '', '2019', '', '2020', '', '2021'], alpha = 0.3)
ax1.text(19800, 1.92, 'TRUMP', fontsize = 18, weight = "bold", color = '#6DA9E4')
ax1.text(19550, 1.80, '(2017 - 2021)', weight = "bold", alpha = 0.3)

# plot all terms

ax4.plot(bush["Time"], bush["rolling_mean"], color = '#ADE4DB')
ax4.plot(obama["Time"], obama["rolling_mean"], color = '#F6BA6F')
ax4.plot(trump["Time"], trump["rolling_mean"], color = '#6DA9E4')
ax4.set_xticks([])


# add title, subtitle

ax1.text(11000, 2.26, 'EURO-USD rate averaged 1.22 under the last three US presidents', fontsize = 20, weight = "bold")
ax1.text(11000, 2.14, 'EURO-USD exchange rates under George W. Bush, Barack Obama, and Donald Trump')

# add signature

ax4.text(10000, 0.65, '@DATAQUEST' + ' '*119 + 'Source: European Central Bank', color = "#f0f0f0", backgroundcolor = "#4d4d4d", size = 14)

plt.show()
```
